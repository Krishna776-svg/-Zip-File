# Question 1 Data Analyst first 10 jobs data.

```python
url = "https://www.naukri.com/data-analyst-jobs-in-bangalore-bengaluru?k=data%20analyst&l=bangalore%2Fbengaluru&cityTypeGid=97"
page= requests.get(url)
page.text
```

```python
driver = webdriver.Chrome("chromedriver.exe")
url="https://www.naukri.com/"
driver.get(url)
```

```python
search_job=driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys("Data Analyst")
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys("Bangalore/Bengaluru")

search_btn=driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()

```

```python
#extract the job title

titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles_tags
job_titles= []
for i in titles_tags:
    job_titles.append(i.text)
job_titles
```

```python
#extract the companies names

companies_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tags
companies_names= []
for i in companies_tags:
    companies_names.append(i.text)
companies_names
```

```python
#extract the experience list

experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
experience_tags
experience_list= []
for i in experience_tags:
    experience_list.append(i.text)
experience_list
```

```python
#extract the locations names

locations_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
locations_tags
locations_names= []
for i in locations_tags:
    locations_names.append(i.text)
locations_names
```

```python
#making dataframe

jobs=pd.DataFrame({})
jobs['title']=job_titles[:10]
jobs['company']=companies_names[:10]
jobs['experience_required']=experience_list[:10]
jobs['location']=locations_names[:10]
jobs
```

# Question 6 Scrape data of first 100 sunglasses listings on flipkart.com

```python
#the webpage 

url = "https://www.flipkart.com/search?q=sunglasses&sid=26x&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=RECENT&suggestionId=sunglasses%7CSunglasses&requestId=2ab638c5-6946-44bf-8e77-2c9e893e391b&as-searchtext=su"
page = requests.get(url)

# content in page1


soup = BeautifulSoup(page.content)
Brand = []  
Product_Description = []    
Price = []   
Discount = []  


# Scrape Brand name
for i in soup.find_all("div",class_="_2WkVRV"):
    Brand.append(i.text)

# Scrape product Description 
for i in soup.find_all("a",class_="IRpwTa"):
    Product_Description.append(i.text)

# scrap price
for i in soup.find_all("div",class_="_30jeq3"):
    Price.append(i.text)
    
# Scrape Discount_%
for i in soup.find_all("div",class_="_3Ay6Sb"):
    Discount.append(i.text)
    
# Make data frame 100 sunglasses listed on flipkart.com
sunglasses=pd.DataFrame({})
sunglasses['Brand']=Brand[:40]
sunglasses['Product_Description']=Product_Description[:40]
sunglasses['Price']=Price[:40]
sunglasses['Discount_%'] = Discount[:40]
sunglasses
```
