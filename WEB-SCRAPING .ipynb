# Python Web Scraping: Extract and display all the header tags from en.wikipedia.org/wiki/Main_Page

```python
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')
bs = BeautifulSoup(html, "html.parser")
titles = bs.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', *titles, sep='\n\n')
```

# python program to extract information about the local weather from the National Weather Service 
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day 

```python
#7days weather info

url = "https://forecast.weather.gov/MapClick.php?lon=-122.49674&lat=37.72964#.YLN2F43itPZ"
r = requests.get(url)
r.content
soup = BeautifulSoup(r.content, "html.parser")

anchors = soup.select("#detailed-forecast-body")
for a in anchors:
    print(a.text)
```

```python
#Extracting all the information from the page

page = requests.get("https://forecast.weather.gov/MapClick.php?lon=-122.49674&lat=37.72964#.YLN2F43itPZ")
soup = BeautifulSoup(page.content, 'html.parser')
seven_day = soup.find(id="seven-day-forecast")
forecast_items = seven_day.find_all(class_="tombstone-container")
tonight = forecast_items[0]
print(tonight.prettify())
```

```python
period = tonight.find(class_="period-name").get_text()
short_desc = tonight.find(class_="short-desc").get_text()
temp = tonight.find(class_="temp").get_text()
print(period)
print(short_desc)
print(temp)
```

```python
img = tonight.find("img")
desc = img['title']
print(desc)
```

```python
period_tags = seven_day.select(".tombstone-container .period-name")
periods = [pt.get_text() for pt in period_tags]
periods
```

```python
short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
descs = [d["title"] for d in seven_day.select(".tombstone-container img")]
```

```python
print(short_descs)
print(temps)
print(descs)
```

```python
import pandas as pd
weather = pd.DataFrame({
    "period": periods,
    "short_desc": short_descs,
    "temp": temps,
    "desc":descs
})
weather
```
